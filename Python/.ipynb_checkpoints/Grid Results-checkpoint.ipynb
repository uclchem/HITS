```python
import uclchem
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
def get_density(history_file):
    df=pd.read_csv(history_file,skiprows=2)
    return df.iloc[-1,1]
```

# Plots

Let's check by eye whether or not things look ok. 

```python
# models=pd.read_csv("models.csv")
# models["density"]=models["history"].map(get_density)
# models.to_csv("models_with_density.csv",index=False)
models=pd.read_csv("models_with_density.csv")
```

```python
models["density"]=np.log10(models["density"]).round(1)
```

```python
unique_phys=models[['radfield', 'zeta', 'density']].drop_duplicates().reset_index(drop=True)
temps=models["initialTemp"].unique()
```

```python
plot_species=["CO","H2O","CH3OH","#CO","HCO+","H","H2"]
for i,row in unique_phys.iloc[0:1].iterrows():
    fig,axes=plt.subplots(3,2,figsize=(8,11),tight_layout=True)
    axes=axes.flatten()
    title_string=""
    idx=models["density"]==row["density"]
    for col,value in row.iteritems():
        idx=idx & (models[col]==value)
        title_string+=f"{col}={value}"
    for j,temp in enumerate(temps):
        plot_idx=idx & (models["initialTemp"]==temp)
        plot_data=pd.DataFrame()
        for k,data_file in enumerate(models.loc[plot_idx,"outputFile"].unique()):
            temp_data=uclchem.read_output_file(data_file)
            temp_data["history"]=f"{k}"
            plot_data=plot_data.append(temp_data[["Time","history"]+plot_species])
        plot_data=plot_data.melt(id_vars=["Time","history"])
        sns.lineplot(data=plot_data,x="Time",y="value",hue="variable",style="history",ax=axes[j],legend="brief")
        axes[j].set(yscale="log",xscale="log",ylim=(1e-12,1),xlim=(1e3,2e6))
        axes[j].text(0.05,0.95,f"{temp} K",va="top",ha="left",transform=axes[j].transAxes)
        if j!=0:
            axes[j].legend().remove()
        else:
            axes[j].legend(ncol=2)
    fig.suptitle(title_string)
    fig.savefig(f"test.png",dpi=300)
```
